```python
import pyttsx3
import datetime
import speech_recognition as sr 
import wikipedia
import webbrowser

engine = pyttsx3.init('sapi5')
#voices = engine.getProperty('voices')
#voices
engine.setProperty('rate', 125) # change the speed of the voice

x = 'Sir. My name is SIVA 1.0. How may I help you?' # SIVA - Super Intelligent Virtual Assistant
########################################################

def wish_me():
    hour = int(datetime.datetime.now().hour)
    
    if hour >=0 and hour < 12:
        engine.say(f"Good morning {x}")
        engine.runAndWait() 
    if hour >=12 and hour <18:
        engine.say(f"Good Evening {x}")
        engine.runAndWait()
    if hour >=18 and hour <24:
        engine.say(f"Good Night {x}")
        engine.runAndWait()
        
########################################################        

def speak(audio):
    engine.say(audio)
    engine.runAndWait()
    
########################################################    

def takeCommand():
    r = sr.Recognizer()
    mic = sr.Microphone()

    with mic as audio_file:
        print("Speak Please")
        r.pause_threshold = 1
        r.adjust_for_ambient_noise(audio_file)
        audio = r.listen(audio_file)
    try:
        print("Listening to the command...")
        task = r.recognize_google(audio, language='en-in')
        print("You said: " + r.recognize_google(audio, language='en-in'))
        
    except Exception as e:
        speak("Couldn't understand what the hell you are saying.")
        return "None"
        
    return task

########################################################
wish_me()
```

```python
query = takeCommand().lower()

if "wikipedia" in query:
    query = query.replace("wikipedia", "")
    query = wikipedia.summary(query, sentences=3)
    speak("According to Wikipedia")
    #print(query)
    speak(query)

if "search" in query and "in youtube" in query:
    query = query.replace("search", "")
    query = query.replace("in youtube", "")
    webbrowser.open(f"https://www.youtube.com/results?search_query={query}")
    
if "search" in query and "in google" in query:
    query = query.replace("search", "")
    query = query.replace("in google", "")
    webbrowser.open(f"https://www.google.com/search?q={query}")
```
